# Get Geocode

```python
#=====================================
#            SETTINGS
#=====================================

#PLEASE WRITE YOUR API CODE INTO YOUR OS ENVIRONMENTAL VARS TO KEEP YOUR API SAFE.
#MY KEY HAS BEEN ENCRYPTED, SO YOU CAN NOT USE IT.
import os
api_key = os.environ['GMAP_API']

#CSVName
fileName = "TrumpSpeech(2017-01-20 to 2018-01-20).csv"
#=====================================

import requests
import pandas as pd

#ReadFile
file = pd.read_csv(fileName)
print('File has been read.')

#Initialize
locationNames = file['Location']
locationList = []
count = 1

#Fetching 
print("Contecting to Google Maps...")
for locationName in locationNames:
    if (locationName != "N/A"):
        url = 'https://maps.googleapis.com/maps/api/geocode/json'
        keys = {'address': locationName, 'key': api_key}
        result = requests.get(url,params=keys)
        result = result.json()
        if (result['status'] == 'ZERO_RESULTS'):
            locationList.append('N/A')
            print('"' + locationName + '" can\'t get the geographical coordinates.\n The result will be replace by "N/A".')
        else:
            coordinates = result['results'][0]['geometry']['location']
            locationList.append([coordinates['lng'], coordinates['lat']])
    else:
        locationList.append('N/A')
    
    #Print the process
    print('{:.0%}'.format(count/len(locationNames)), "completed.") 
    count += 1

print("Done. Outputing CSV now...")

#Output CSV
total = dict(zip(
    ['Date', 
     'Title',
     'Location',
     'Coordinates',
     'Link',
     'Speech'],
    [file['Date'], 
     file['Title'],
     file['Location'],
     locationList,
     file['Link'],
     file['Speech']]))

table = pd.DataFrame(total)
table.head()
table.to_csv(fileName.replace('.csv', '_Located') + ".csv", index=False)
print("The CSV has been output.")
```
