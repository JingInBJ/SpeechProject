# How many speeches has Donald Trump done?

```python
#set a start date
s_date = "2017-01-20"
#set a end date
e_date = "2018-01-20"
#====================================

import requests
import math
import time
import pandas as pd
from bs4 import BeautifulSoup, element
import time
import urllib3

#Disable warnings
urllib3.disable_warnings()

def dateTime (d):
    return time.strptime(d, "%Y-%m-%d")

#Initialize
dateList = []
titleList = []
linkList = []
locationList = []
speechContent = []

#Count the num of records, and calculate pages
result = requests.get("https://factba.se/json/json-transcript.php?q=&p=&f=s&dt=2019-06-02")
records = result.json()
pages = math.ceil(int(records['meta']['total_rows'])/len(records['data']))

#Initialize counter
x = 1

#Search pages
while x <= pages:
    #URL definition
    url = "https://factba.se/json/json-transcript.php?p=" + str(x) + "&f=s"
    results = requests.get(url).json()
    results = results['data']
    
    for result in results:
        #If the date under the rule
        if (dateTime(result['date']) >= dateTime(s_date) and dateTime(result['date']) <= dateTime(e_date)):
            #Add dates and titles into the lists
            dateList.append(result['date'])
            titleList.append(result['record_title'])
            
            #If location info is missing
            if (result['city'] != None and result['state_code'] != None and result['country'] != None):
                locationList.append(result['city'] + ', ' + result['state_code'] + ', ' + result['country'])
            elif (result['city'] == None and result['state_code'] == None and result['country'] == None):
                locationList.append("N/A")
            else:
                locationList.append(result['country'])
            #Add links
            linkList.append(result['slug'])
     
    #Print process
    print('{:.0%}'.format(x/pages), "completed.") 
    #Page counter
    x += 1

print("Scraping contexts, please wait...")
#Scrap all contexts
for link in linkList:
    url = "https://factba.se/transcript/" + link
    response = requests.get(url, verify=False)
    doc = BeautifulSoup(response.text)
    
    contexts = doc.find_all(class_='transcript-text-block')
    speech = ''
    speech += '\n'.join([context.text.strip() for context in contexts])
    
    #Write into the lists
    speechContent.append(speech)

print("Done. \n")
fileName = input("Please input CSV file name for output. \n")

#Output CSV
total = dict(zip(
    ['Date', 
     'Title',
     'Location',
     'Link',
     'Speech'],
    [dateList, 
     titleList,
     locationList,
     linkList,
     speechContent]))

table = pd.DataFrame(total)
table.head()
table.to_csv(fileName + "(" + s_date + " to " + e_date + ").csv", index=False)
print("The CSV has been output.")
```
